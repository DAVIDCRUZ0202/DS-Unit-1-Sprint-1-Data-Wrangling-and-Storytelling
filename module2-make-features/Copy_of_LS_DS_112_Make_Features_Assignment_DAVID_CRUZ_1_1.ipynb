```python
# Make a column named loan_status_is_great. It should contain the integer 1 if
#  loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

def loan_fix(variable_loan):

  if variable_loan == "Current" or variable_loan == "Fully Paid":
    return 1
  else:
   return 0



df['loan_status_is_great'] = df['loan_status'].apply(loan_fix)

df.head(100)
```
